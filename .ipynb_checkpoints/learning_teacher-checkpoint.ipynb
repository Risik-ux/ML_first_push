### 2.3. Анализ дубликатов.

#### 2.3.1. Поиск полных дублей.

```python
# функция поиска и удаления явных дубликатов, в процессе выводится справочния информация
def custom_drop_duplicates(df):
    dup_count = df.duplicated().sum()
    if dup_count:
        tmp_shape = df.shape
        df = df.drop_duplicates().reset_index(drop=True)
        print(
            f'Выявлено и удалено {dup_count} явных дубликатов.\n'
            f'Размер датафрейма до изменений {tmp_shape[0]}.\n'
            f'Размер датафрейма после изменений {df.shape[0]}.\n'
            f'Процент дубликатов {1-(df.shape[0]/tmp_shape[0]):.2%}'
        )
    else:
        print(
            f'Полных дубликатов не выявлено.\n'
            f'Размер датафрейма остался {df.shape[0]}.\n'
        )
    return(df)
```

```python
market_main = custom_drop_duplicates(market_main)
```

```python
market_money = custom_drop_duplicates(market_money)
```

```python
market_time = custom_drop_duplicates(market_time)
```

```python
money = custom_drop_duplicates(money)
```

```python
def dup_check(*arg):
    for index, df in enumerate(arg):
        print(f'\nАнализ категорийный признаков датафрейма №{index+1}:')
        columns = df.select_dtypes(include='object').columns
        if len(columns):
            for column in columns:
                print(f'Признак: {column}\nЗначения: {df[column].unique()}')
        else:
            print('В датафрейме нет категорийных признаков.')
```

### 3.2. Визуализация распределения количественных признаков.

```python
# кастомная функция формирования фигуры для подграфов
def custom_subplot(plot_settings):
    return(
        plt.subplots(
            nrows=plot_settings[0],
            ncols=plot_settings[1],
            figsize=(
                plot_settings[2],
                plot_settings[3]
            )
        )
    )
```

```python
# функция отрисовки подграфа типа box
def custom_boxplot(axx, data, label):
    sns.boxplot(
        ax=axx,
        data=data,
        showmeans=True,
        color=COLORS[random.randrange(7)]
    ).set(
        xlabel=label, 
        ylabel=f'Значение параметра {label}'
    )
```

```python
# функция для отрисовки подграфа типа hist
def custom_hist(axx, data, label):
    sns.histplot(
        ax=axx,
        data=data,
        color=COLORS[random.randrange(7)],
        kde=True,
        bins=50
    ).set(
        xlabel=label, 
        ylabel='Количество'
    )
```

```python
# кастомная функция для создания графов в одной фигуре
def custom_graph(df, columns, plot_settings, title, kind):
    fig, ax = custom_subplot(plot_settings)
    fig.suptitle(title, fontsize=20)
    index=0
    for i in range(plot_settings[0]):
        for j in range(plot_settings[1]):
            try:
                axx = ax[i,j] # если в фигуре создан один ряд
            except:           # то возникнет исключение
                axx = ax[j]   # делаем перебор только по столбцам
            try:
                if kind == 'box':
                    custom_boxplot(axx, df[columns[index]], columns[index])
                elif kind == 'hist':
                    custom_hist(axx, df[columns[index]], columns[index])
                else:
                    print(f'Попытка вызвать неописанный тип графа - {kind}')
                    return
            except IndexError:
                fig.delaxes(axx)
                return
            index += 1
    plt.show()
```

```python
# определим параметры графов для датафрейма market_main
# столбцы по которым будем строить диаграммы
# main_columns = market_main.drop(['id'], axis=1).select_dtypes(exclude='object').columns.to_list()
main_columns = ['Маркет_актив_6_мес', 'Длительность', 'Акционные_покупки']
# определим столбцы с дискретным распределением
main_columns_dif = ['Маркет_актив_тек_мес', 'Средний_просмотр_категорий_за_визит', 
                    'Неоплаченные_продукты_штук_квартал', 'Ошибка_сервиса', 'Страниц_за_визит']
# параметры фигуры для постройки количество строк, столбцов, ширина, высота
main_graph_settings = [1, 3, 15, 5]
# подпись фигуры
main_title = 'Диаграммы распределения непрерывных признаков датафрейма market_main'
```

```python
custom_graph(market_main, main_columns, main_graph_settings, main_title, 'box')
```

В датафрейме `market_money` у нас есть анамалии в виде нулевой выручки и выручки более 100.000.
* завышеную выручку мы заменим на медианное значение.
* пользователей с нулевой выручкой удалим.

```python
market_money.loc[market_money['Выручка'] > 100_000, 'Выручка'] = market_money['Выручка'].median()
```

```python
# подготовим список пользователей с нулевой выручкой
zero_ids = market_money[market_money['Выручка'] == 0]['id'].unique().tolist()
print(
    f'Пользователей с нулевой выручкой {len(zero_ids)} человека\n'
    f'Их ID: {zero_ids}'
)
```

```python
def user_del(*dfs):
    res=[]
    for index, df in enumerate(dfs):
        start_shape = df.shape[0]
        df = df[~df['id'].isin(zero_ids)].copy()
        print(f'В датафрейме {index+1} удалено {start_shape - df.shape[0]} записей.')
        res.append(df)
    return(res)
```

```python
market_main, market_money, market_time, money = user_del(market_main, market_money, market_time, money)
```

## 4. Объединение таблиц.

```python
# Посмотрим размерность датпфремов
print(f'Размер market_main: {market_main.shape}')
print(f'Размер market_money: {market_money.shape}')
print(f'Размер market_time: {market_time.shape}')
print(f'Размер money: {money.shape}')
```

Исходный датафрейм имеет `13` столбцов. Необходимо объединить его с датафреймами `market_money` и `market_time`. Датафрейм `market_money` содержит `3891` записей по `3` периода на каждого пользователя. Необходимо трансформировать его, чтобы название периодов стали столбцами. Датафрейм `market_time` содержит `2594` записей по `2` периода на пользователя. Его так же надо трансформировать. Итого получается должно добавиться `5` столбцов, `2` периода времени и `3` периода выручки.

```python
# функция трансформирует датафрейм и меняет название столбцов
def df_transform(df, column, prefix):
    df_mod = df.pivot_table(index='id', columns=column)
    col_names=[]
    for col in df_mod.columns:
        col_names.append(col[1])
    df_mod.columns = col_names
    return(df_mod.add_prefix(prefix))
```

```python
# объединяем три датафрейма в один
market_full = market_main.join(
    df_transform(market_money, 'Период', 'Выручка_'), on='id'
).join(
    df_transform(market_time, 'Период', 'Время_'), on='id'
)
# выводим размерность исходного датафрейма и несколько строк с данными
print(market_full.shape)
market_full.sample(3)
```

Количество строк не изменилось, количество столбцов увеличилось на 5, как и предполагалось.

```python
# признаки для OHE кодирования
ohe_cols = ['Разрешить_сообщать', 'Популярная_категория']
```

```python
# признаки для Ordinal кодирования
ord_cols = ['Тип_сервиса']
```

```python
# количественные признаки для маштабирования
num_cols = market_full.drop(TARGET_ATTRIBUTE, axis=1).drop('id', axis=1).select_dtypes(exclude='object').columns.to_list()
```

```python
# пайплайн для OHE кодирования признаков. пропуски заменям модой.
ohe_pipe = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False))
    ]
)
```

```python
# пайплайн для Ordinal кодирования признаков. пропуски заменям модой.
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум'], 
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_cols),
        ('ord', ord_pipe, ord_cols),
        ('num', StandardScaler(), num_cols)
    ], 
    remainder='passthrough'
) 
```

```python
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
) 
```

```python
param_grid = [
    # параметры KNeighborsClassifier
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },
    # параметры DecisionTreeClassifier
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # параметры LogisticRegression
    {
        'models': [
            LogisticRegression(
                random_state=RANDOM_STATE,
                solver='liblinear',
                penalty='l1'
            )
        ],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # параметры SVC
    {
        'models': [
            SVC(random_state=RANDOM_STATE, kernel='poly', probability=True)
        ],
        'models__degree': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

```python
random_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring='recall',#'roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```
